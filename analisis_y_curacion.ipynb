```python
import pandas
drs = pandas.read_csv('https://raw.githubusercontent.com/alemaldonado/diplodatos/master/dirty_retail_sales.csv')
drs.info()
```

```python
drs.shape
```

```python
drs
```

### 1

```python
drs['Purchase'] = drs.Purchase.astype(float)/100
drs.head()
```

```python
drs.info()
```

### 2

```python
drs = drs.join(drs['Gender_Age_Marital_Status'].str.split('_', 2, expand=True).rename(columns={0: 'Gender', 1: 'Age', 2: 'Marital_Status'}))
drs = drs.drop(columns=['Gender_Age_Marital_Status'])
drs.head()
```

```python
drs.info()
```

### 3

```python
drs_grouped_by_product = drs['Purchase'].groupby(drs["Product_ID"])
drs_grouped_by_product.describe()
```

### 4

```python
new_purchase_values = round(drs_grouped_by_product.mean(),2).to_frame().reset_index()
drs = drs.merge(new_purchase_values, how='left', on='Product_ID').drop(columns=['Purchase_x']).rename(columns={'Purchase_y':'Purchase'})
drs
```

```python
drs['Purchase'].groupby(drs["Product_ID"]).describe().head()
```

```python
grouped_users = drs.groupby(['User_ID', 'Gender', 'Age', 'Occupation', 'City_Category', 'Stay_In_Current_City_Years', 'Marital_Status']).size().reset_index(name='Frequency')
grouped_users
```

```python
grouped_users[grouped_users.User_ID.duplicated(keep=False)]
```

```python
grouped_products = drs.groupby(['Product_ID', 'Product_Category_1', 'Product_Category_2', 'Product_Category_3']).size().reset_index(name='Frequency')
grouped_products
```

```python
grouped_products[grouped_products.Product_ID.duplicated(keep=False)]
```

### 6

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
city_category_le = LabelEncoder()
city_category_labels = city_category_le.fit_transform(drs['City_Category'])
drs['City_Category_Label'] = city_category_labels
age_le = LabelEncoder()
age_labels = age_le.fit_transform(drs['Age'])
drs['Age_Label'] = age_labels
drs_sub = drs[['User_ID', 'City_Category', 'City_Category_Label', 'Age', 'Age_Label']]
drs_sub.head(20)
```

```python
city_category_ohe = OneHotEncoder()
city_category_feature_arr = city_category_ohe.fit_transform(drs[['City_Category_Label']]).toarray()
city_category_feature_labels = list(city_category_le.classes_)
city_category_features = pandas.DataFrame(city_category_feature_arr, columns=city_category_feature_labels)

age_ohe = OneHotEncoder()
age_feature_arr = age_ohe.fit_transform(drs[['Age_Label']]).toarray()
age_feature_labels = list(age_le.classes_)
age_features = pandas.DataFrame(age_feature_arr, columns=age_feature_labels)

drs_ohe = pandas.concat([drs_sub, city_category_features, age_features], axis=1)
columns = sum([['User_ID', 'City_Category', 'City_Category_Label'], city_category_feature_labels, ['Age', 'Age_Label'], age_feature_labels], [])
drs_ohe[columns].head(20)
```
